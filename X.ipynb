#PRE PROCESSING & CLEANING TEXT

##Import Library

```python
import re
import string
```

```python
df = df_rename.copy()
display(df[['TWEET']].head(10))
```

##Lowercase

```python
df['TWEET_lower'] = df['TWEET'].str.lower()
display(df[['TWEET', 'TWEET_lower']].head(5))
```

##Hapus URL

```python
df['TWEET_no_url'] = df['TWEET_lower'].apply(lambda x: re.sub(r'http\S+', '', x))
display(df[['TWEET_lower', 'TWEET_no_url']].head(15))
```

##Hapus Mention

```python
df['TWEET_no_mention'] = df['TWEET_no_url'].apply(lambda x: re.sub(r'@\w+', '', x))
display(df[['TWEET_no_url', 'TWEET_no_mention']].head(5))
```

##Hapus Hastag

```python
df['TWEET_no_hashtag'] = df['TWEET_no_mention'].apply(lambda x: re.sub(r'#\w+', '', x))
display(df[['TWEET_no_mention', 'TWEET_no_hashtag']].head(15))
```

Hapus karakter jepang

```python
import re

# Convert list → string (jika perlu)
df['TWEET_clean_str'] = df['TWEET_no_hashtag'].apply(
    lambda x: ' '.join(x) if isinstance(x, list) else str(x)
)

# 1. Hapus karakter Jepang (Hiragana, Katakana, Kanji)
df['no_japan'] = df['TWEET_clean_str'].str.replace(
    r'[\u3040-\u309F\u30A0-\u30FF\u4E00-\u9FFF]',
    ' ',
    regex=True
)

# 2. Hapus bracket Jepang
df['no_japan'] = df['no_japan'].str.replace(
    r'[『』「」【】《》〈〉]',
    ' ',
    regex=True
)

# 3. Rapikan spasi
df['no_japan'] = df['no_japan'].str.replace(r'\s+', ' ', regex=True).str.strip()

# Hasil final:
df[['TWEET_no_hashtag', 'no_japan']]
```

##Hapus tanda baca tetapi tetap memisahkan kata

```python
punctuation_pattern = "[" + re.escape(string.punctuation) + "]"
df['TWEET_no_punct'] = df['TWEET_no_hashtag'].apply(lambda x: re.sub(punctuation_pattern, " ", x))
df['TWEET_no_punct'] = df['TWEET_no_punct'].str.replace(r'\s+', ' ', regex=True).str.strip()
display(df[['TWEET_no_hashtag', 'TWEET_no_punct']].head(5))
```

Normalisasi inggris indo

```python
import pandas as pd
kamus_df = pd.read_csv('kamus_inggris_x.csv', on_bad_lines='skip')
kamus_df['Inggris'] = kamus_df['Inggris'].astype(str).str.strip()
kamus_df['Indonesia'] = kamus_df['Indonesia'].astype(str).str.strip()
kamus_df['Indonesia'] = kamus_df['Indonesia'].fillna("")
kamus_manual = dict(zip(kamus_df['Inggris'], kamus_df['Indonesia']))
kamus_manual
```

```python
def translate_manual(text):
    if pd.isna(text):
        return ""

    words = text.split()
    translated = []
    for w in words:
        if w in kamus_manual:
            translated.append(kamus_manual[w])
        else:
            translated.append(w)
    return " ".join(translated)
```

```python
df['TWEET_clean'] = df['TWEET_clean'].fillna("")
df['Comment_indo'] = df['TWEET_clean'].apply(translate_manual)
df[['TWEET_clean', 'Comment_indo']].head(5)
```

##Normalisasi

```python
norm_df = pd.read_csv('normalisasi_manual_x.csv')
norm_manual = dict(zip(norm_df['tidak baku'], norm_df['baku']))
norm_manual
```

```python
def normalisasi_manual(text):
    if not isinstance(text, str):
        return text

    words = text.lower().split()
    norm = []

    for w in words:
        w_str = str(w)
        if w_str in norm_manual:
            norm.append(norm_manual[w_str])
        else:
            norm.append(w_str)

    return " ".join([str(n) for n in norm])
```

```python
df['Comment_norm'] = df['Comment_indo'].apply(normalisasi_manual)
df[['Comment_indo', 'Comment_norm']].head(5)
```

```python
import re

def remove_repeated_full_words(text):
    # hapus kata yang sama muncul berurutan 2x atau lebih
    return re.sub(r'\b(\w+)( \1\b)+', r'\1', text, flags=re.IGNORECASE)

df["Comment_norepeatword"] = df["Comment_norm"].apply(remove_repeated_full_words)
df[["Comment_norm", "Comment_norepeatword"]]
```

##Hapus stopwords (dan, di, yang, untuk, saja, ke)

```python
%pip install Sastrawi
```

```python
from Sastrawi.StopWordRemover.StopWordRemoverFactory import StopWordRemoverFactory
```

```python
factory = StopWordRemoverFactory()
stopword = factory.get_stop_words()

# Stopword tambahan (custom)
custom_stopwords = [
    "dan", "dah", "s", "ini", "nan", "oleh", "saya", "oleh", "sih", "kamu", "nik", "mingu", "katseye",
    "situ", "sana", "dong", "kan", "apa", "nya", "meni", "hehe", "dar", "embut", "jadi", "kok", "heso",
    "banget", "t", "aku", "enggak", "loh", "saya", "saw", "wah", "kalau", "elu", "tuh", "kak"
]

# Gabungkan stopword bawaan + custom
all_stopwords = set(stopword + custom_stopwords)

def remove_stopwords(data):
    if not isinstance(data, str):
        return []
    words = data.split()
    filtered = [w for w in words if w not in all_stopwords]
    return filtered

df['TWEET_clean'] = df['Comment_norepeatword'].apply(lambda x: remove_stopwords(x))
df[['Comment_norepeatword', 'TWEET_clean']]
```

## Stemmed

```python
from Sastrawi.Stemmer.StemmerFactory import StemmerFactory
factory = StemmerFactory()
stemmer = factory.create_stemmer()
df['Comment_stemmed'] = df['TWEET_clean'].apply(lambda x: stemmer.stem(' '.join(x)) if isinstance(x, list) else stemmer.stem(str(x)))
df[['TWEET_clean', 'Comment_stemmed']]
```

```python
df['clean_for_wc'] = df['Comment_stemmed']

# 1. Hapus huruf berulang
df['clean_for_wc'] = df['clean_for_wc'].str.replace(r'(.)\1{2,}', r'\1', regex=True)

# 2. Hapus semua karakter non ASCII
df['clean_for_wc'] = df['clean_for_wc'].str.replace(r'[^\x00-\x7F]', ' ', regex=True)

# 3. Hapus semua tanda baca
df['clean_for_wc'] = df['clean_for_wc'].str.replace(r'[^\w\s]', ' ', regex=True)

# 4. Rapikan spasi
df['clean_for_wc'] = df['clean_for_wc'].str.replace(r'\s+', ' ', regex=True).str.strip()

# Handle potential non-string (e.g., float/NaN) values before splitting
df['dedup'] = df['clean_for_wc'].apply(
    lambda x: " ".join(dict.fromkeys(x.split())) if pd.notna(x) and isinstance(x, str) else ""
)
```

Analisis N-Grams

```python
from sklearn.feature_extraction.text import CountVectorizer

def get_top_ngrams(corpus, n=None, top_k=10):
    vec = CountVectorizer(ngram_range=(n, n)).fit(corpus)
    bag_of_words = vec.transform(corpus)
    sum_words = bag_of_words.sum(axis=0)
    words_freq = [(word, sum_words[0, idx]) for word, idx in vec.vocabulary_.items()]
    words_freq = sorted(words_freq, key = lambda x: x[1], reverse=True)
    return words_freq[:top_k]

top_bigrams = get_top_ngrams(df[col_name].astype(str), n=2, top_k=5)
df_bigram = pd.DataFrame(top_bigrams, columns=['Bigram', 'Frekuensi'])

plt.figure(figsize=(12, 6))
sns.barplot(x='Frekuensi', y='Bigram', data=df_bigram, palette='magma')
plt.title('5 Frasa (2 Kata) Paling Sering Muncul')
plt.show()
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.svm import LinearSVC
from sklearn.metrics import accuracy_score, classification_report
```

```python
df = pd.read_csv("X_comments_label manual.csv")
text_column = df['dedup']
label_column = df['hf_label']
```

```python
X = text_column
Y = label_column

X_train, X_test, Y_train, Y_test = train_test_split(
    X, Y, test_size=0.2, random_state=42
)
mask = ~pd.Series(Y_train).isna()
X_train = X_train[mask].fillna('')
Y_train = pd.Series(Y_train)[mask]
X_test = X_test.fillna('')
```

# TF IDF

```python
tfidf = TfidfVectorizer(
    max_features=5000,
    ngram_range=(1,2)
)

X_train_tfidf = tfidf.fit_transform(X_train.fillna(''))
X_test_tfidf = tfidf.transform(X_test.fillna(''))
```

```python
print(X_train_tfidf.toarray())
```

10 kata tertinggi

```python
tfidf = TfidfVectorizer(
    ngram_range=(1, 1)
)

X_train_tfidf = tfidf.fit_transform(X_train)
```

```python
import numpy as np

tfidf_sum = X_train_tfidf.sum(axis=0).A1
feature_names = tfidf.get_feature_names_out()

top_idx = np.argsort(tfidf_sum)[::-1][:10]
top_features = [(feature_names[i], tfidf_sum[i]) for i in top_idx]

print("\nTop 10 kata (berdasarkan TF-IDF total):")
for f, v in top_features:
    print(f"{f}: {v:.4f}")
```

```python
# hitung rata2 tf idf per fitur
mean_tfidf = np.mean(X_train_tfidf, axis=0).A1
feature_names = tfidf.get_feature_names_out()

top_n = 10
top_idx = mean_tfidf.argsort()[::-1][:top_n]

top_features = feature_names[top_idx]
top_scores = mean_tfidf[top_idx]

plt.figure(figsize=(10, 8))
plt.barh(top_features[::-1], top_scores[::-1])
plt.title("Top 10 TF-IDF Features (Train Set)")
plt.xlabel("Average TF-IDF Score")
plt.ylabel("Feature")
plt.tight_layout()
plt.show()
```

#MNB

```python
X_train_tfidf = tfidf.fit_transform(X_train.fillna(''))
X_test_tfidf = tfidf.transform(X_test.fillna(''))

mnb = MultinomialNB()
mnb.fit(X_train_tfidf, Y_train)
mnb_pred = mnb.predict(X_test_tfidf)

print("\n= Multinomial Naive Bayes =")
print("Accuracy:", accuracy_score(Y_test, mnb_pred))
print(classification_report(Y_test, mnb_pred))
```

```python
from sklearn.metrics import confusion_matrix

cm = confusion_matrix(Y_test, mnb_pred)
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
plt.title("Confusion Matrix - MNB")
plt.show()
```
